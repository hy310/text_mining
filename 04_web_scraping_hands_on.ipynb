### <1> 전자신문 사이트에서 '인공지능' 검색

#### 필요 라이브러리 import

```python
## 기본 패키지 불러오기
import numpy as np 
import pandas as pd
import requests
import warnings
warnings.filterwarnings('ignore')
```

#### status_code 이용 에러여부 확인

```python
res = requests.get('https://search.etnews.com/etnews/search.php?kwd=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5')

print("응답 코드: ", res.status_code)  # 응답코드가 200 이면 정상,  그렇지 않으면 에러가 발생한 경우 (400번대 코드)
print('\n')

if res.status_code ==  requests.codes.ok :
    print("정상입니다")
else:
    print("에러가 발생했습니다. [에러코드 ", res.status_code, "]")
```

#### params option 작성

```python
# 전자신문 사이트에서 '인공지능' 검색

url = 'https://search.etnews.com/etnews/search.php'

params = { 
    "kwd": "인공지능"
}

res = requests.get(url, params=params)
res.text
```

#### 데이터 수집 후 해당 사이트 인코딩 방식을 확인

```python
res.encoding
```

#### 수집된 데이터 사이즈 확인 및 데이터 수집 후 내용 확인

```python
res.content # 데이터 사이즈 확인
```

```python
print(len(res.text))   # 수집된 데이터 크기 확인
```

```python
print(res.text)      # 수집된 내용을 확인
```
