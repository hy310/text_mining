```python
# ë‘ìŒì ˆ ì´ìƒë§Œ ì¶”ì¶œ


okt = Okt()

def tokenizer(text) :
    nouns = okt.@@@(text)
    return [noun for noun in nouns if len(noun) > 1]
```
